```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVC
from sklearn import metrics
```

```python
data = pd.read_csv("/content/netflix_titles_nov_2019.csv", sep=",")
```

```python
data.head()
```

```python
#data = data[data['type'] == 'Movie']

def cambiar_valor_columna_type(season_str):
  if 'Season' in season_str:
    num = int(season_str.split()[0])
    return f'{num * 12 * 40} min' #12 caps * 40 min
  return season_str
data['duration'] = data['duration'].apply(cambiar_valor_columna_type)
data
```

```python
data['duration'] = data['duration'].str.replace(' min', '').astype(int)
data['release_year'] = data['release_year'].astype(int)
data = data.drop(['title', 'description', 'show_id'], axis=1)
data
```

```python
encoderGender=LabelEncoder()
data['type']=encoderGender.fit_transform(data['type'])
data['country']=encoderGender.fit_transform(data['country'])
data['listed_in']=encoderGender.fit_transform(data['listed_in'])
data['rating']=encoderGender.fit_transform(data['rating'])
data['director']=encoderGender.fit_transform(data['rating'])
data['cast']=encoderGender.fit_transform(data['rating'])
data
```

```python
data['date_added'] = data['date_added'].fillna('November 1, 2019')
data
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], format="%B %d, %Y").dt.date
data['date_added'] = pd.to_datetime(data['date_added'])
fecha_minima = data['date_added'].min()
fecha_maxima = data['date_added'].max()
data['dias_desde_inicio'] = data['date_added'] - fecha_minima
data['dias_desde_fin'] = fecha_maxima - data['date_added']
data['dias_desde_inicio'] = data['dias_desde_inicio'].dt.days
data['dias_desde_fin'] = data['dias_desde_fin'].dt.days
print(fecha_minima, fecha_maxima)
data
```

```python
X = data.drop(['duration', 'date_added', 'dias_desde_fin'], axis=1)
y = data['duration']
X
```

```python
grado = 2
poly = PolynomialFeatures(degree=grado)
X_poly = poly.fit_transform(X)
```

```python
modelo = LinearRegression()
modelo.fit(X_poly, y)
```

```python
y_pred = modelo.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Coeficientes:", modelo.coef_)
print("TÃ©rmino independiente:", modelo.intercept_)
print("Error cuadrÃ¡tico medio (MSE):", mse)
print("Coeficiente de determinaciÃ³n (RÂ²):", r2)
```
